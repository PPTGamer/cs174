1. Write a function to check whether a string is a palindrome. (A palindrome is a string that reads the same forwards and backwards) **1 pt**.

We assume all inputs are in lowercase. 

```python
def isPalindrome(str):
    for i in range(len(str)//2 + 1):
        if str[i] != str[len(str)-i-1]:
            return False
    return True
```

```python
print(isPalindrome('abba'))
print(isPalindrome('kayak'))
print(isPalindrome('racecar'))
print(isPalindrome('facecar'))
print(isPalindrome('rakecar'))
print(isPalindrome('abca'))
```

2. Write a function that checks whether a number is prime or not **1 pt.**

```python
def isPrime(x):
    if x == 0 or x == 1: 
        return False
    d = 2
    while d*d < x:
        if x % d == 0:
            return False
        d += 1
    return True
```

```python
for i in range(20):
    print(i, isPrime(i))
```

3. Load `data/allstar.txt` and create a dictionary where the keys are the verses or chorus numbers, and the values are the chords for that part of the song. **4 pts.** `{ "verse1": "Am, C, D, G, F, G, G, Bm", "chorus1": "C, Am, G, D, C, Am, G, D, C"..}`

```python
f = open('data/allstar.txt', 'r')
allstar = f.readlines()
f.close()
valid_chords = []
for k in ['A','B','C','D','E','F','G']:
    valid_chords.append(k)
    valid_chords.append(k+'#')
    valid_chords.append(k+'m')
    valid_chords.append(k+'#m')
```

```python
chords = {}
verse_number, chorus_number = 1, 1
current_section = ''
for line in allstar:
    # new section
    if line[0] == '[':
        section_name = line[1:-1].replace(' ','').lower()
        if 'verse' in section_name:
            current_section = 'verse' + str(verse_number)
            verse_number += 1
        elif 'chorus' in section_name:
            current_section = 'chorus' + str(chorus_number)
            chorus_number += 1
    else:
        chordlist = list(filter(lambda x: x in valid_chords,line.split()))
        if current_section in chords:
            chords[current_section] += chordlist
        else:
            chords[current_section] = chordlist

for section in chords:
    chords[section] = ', '.join(chords[section])
print(chords)
```

4. Build a dictionary from `data/allstar.txt` where the keys are the unique words, and the values are the number of times the word appeared in the song. **4 pts.**

```python
# filter out chords and section headers to form the wordlist
wordlist = []
for line in allstar:
    if line[0] != '[':
        wordlist += list(map(lambda x: x.lower(), filter(lambda x: x not in valid_chords,line.split())))
print(len(wordlist), 'words')
print('wordlist preview:', wordlist[0:10])

def getFrequencyDict(wordlist):
    frequency_dict = {}
    for word in wordlist:
        if word in frequency_dict:
            frequency_dict[word] += 1
        else:
            frequency_dict[word] = 1
    return frequency_dict

allstar_frequency = getFrequencyDict(wordlist)
print(len(allstar_frequency), 'unique words')
print(allstar_frequency)
```

Observation: the lyrics are all found in a plain `<div>` (i.e. no attributes in the tag). We can thus look for this `<div>` and its matching `</div>` to extract the lyrics.

```python
f = open('data/Rick Astley - Never Gonna Give You Up Lyrics _ AZLyrics.com.html', 'r')
neverdata = f.readlines()
f.close()

lyricsstart = neverdata.index('<div>\n')
lyricsend = neverdata.index('</div>\n', lyricsstart)

# Ignore comment at start of lyrics and remove <br> tag and newline.
neverlyrics = [x.replace('<br>','').replace('\n','') for x in neverdata[lyricsstart+2:lyricsend]]
neverlyrics[:10]
```

5a. How many unique words are there?

```python
import string
def getWordList(lyrics):
    wordlist = []
    for line in lyrics:
        # split to remove whitespace.
        for word in line.split():
            # transform to lowercase, remove leading/trailing punctuation.
            word = word.lower().strip(string.punctuation)
            wordlist.append(word)
    return wordlist

never_wordlist = getWordList(neverlyrics)
never_wordset = set(never_wordlist)

print('There are', len(never_wordset), 'unique words.')
print(sorted(never_wordset))
```

5d. Use the dictionary from above to create a word count histogram.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def plotFreqDict(frequency_dict):
    words, freqs = zip(*frequency_dict.items()) 

    indSort = np.argsort(freqs)
    words = np.array(words)[indSort]
    freqs = np.array(freqs)[indSort]

    plt.barh(range(len(frequency_dict)), freqs)
    plt.ylim(-0.5,len(words))
    plt.yticks(range(len(words)), words)
    plt.xlabel('Frequency')
    plt.plot()
    
plt.rcParams['figure.figsize'] = [10, 20]
plt.title('Word count histogram of Rick Astley\'s "Never Gonna Give You Up"')
plotFreqDict(never_frequency_dict)
plt.show()
```

#### Observations
**Both word frequency histograms display a similar shape.**
- A handful of words are used extremely frequently, while a majority of words are used infrequently. There is a clear "curve" in the data among the most frequently used words. The curves look similar, despite the songs having different word counts and being in different languages.
- The overall shape of the distribution is almost like an exponential curve, similar to that predicted by Zipf's law, although this was not verified mathematically for this data.

**There is a higher density of unique words in the Japanese song than in the English song.**  
- Japanese grammar permits a wider variety of verb conjugations than in English. English verb conjugations mainly reflect tense, for example, notice how `go` and `going` are counted separately in the word count histogram. However, Japanese verb conjugations can reflect a wider variety of meanings. Consider the following conjugations of the verbs `ikiru` and `shinu` which are all found in the Japanese song. These are all considered as seperate words (since we have not done lemmatization yet)
    
|ikiru  | root word (to live)                                  |
|-------|------------------------------------------------------|
|ikite  | connective form                                      |
|ikiteku| (contraction of ikiteiku) to go and live / to live on|
|ikiro  | imperative (be alive!)                               |
|ikiyou | volitional (let's live)                              |

|shinu       | root word (to die)                   |
|------------|--------------------------------------|
|shinitai    | to want to die                       |
|shinde      | connective form                            |
|shindara    | conditional form (if (subject) were to die)|
|shinazu     | to do something without dying        |
|shinitakunai| to not want to die                   |

- These verb conjugations also explain why Japanese has a higher average word length than English. The average word length of all unique words in the Japanese song is `5.622950819672131` while the average word length of all unique words in the English song is `4.462962962962963`.

**In the English song, prepositions and conjunctions such as 'to', 'and', 'or', 'on' 'in' have varying frequencies. The approximate equivalent Japanese concept (particles) such as 'ni', 'mo', 'wa', 'ga', 'o', 'no' have generally higher frequencies.** 
- Notice how the 5 most frequently used words in the Japanese songs are all particles, while in the English song, 'time' and 'i' are not prepositions nor conjunctions, but are among the 5 most freqently used words in the English song. This is due to how particles in Japanese are necessary to indicate parts of speech such as the subject, location, and direct object, while in English, parts of speech such as the subject and direct object are determined by the noun's placement with respect to the main verb in the sentence.

**Linking verbs such as 'is' and 'be' in the English song are more frequent than the copulas 'desu', 'datta'.**
- In the English song, 'is' and 'be' both have frequency 6, while 'desu' and 'datta' occur only once each in the Japanese song.
- This is likely due to how in English, certain verb forms still require a linking verb, such as in `the train is leaving` and `time is wasting`, while the copula is absent in Japanese when a verb is used instead.
